# Reading in the Data

```python
import pandas as pd
import numpy as np
```

```python
# Reading in the data
train_data = pd.read_csv("~/Desktop/my_projects/data/home_data/train.csv")
train_data
```

```python
test_data = pd.read_csv("~/Desktop/my_projects/data/home_data/test.csv")
test_data
```

# Cleaning the Training Data

```python
# Looking at the training data
train_data.describe()
```

```python
# Looking at missing values
train_missing_values = train_data.isnull().sum().sort_values(ascending=False)
print(train_missing_values.to_string())
```

```python
# Calculating the percentage of missing values 
train_missing_values_pct = train_data.isnull().sum() * 100 / len(train_data)
train_missing_values_pct = train_missing_values_pct.sort_values(ascending=False)
print(train_missing_values_pct.to_string()) # to_string() to print all values 
```

```python
# Replacing missing values of FireplaceQu with 0s
train_data['FireplaceQu'] = train_data['FireplaceQu'].fillna(0) # Replacing with 0 since Fireplaces with 0 values are just as many as the missing values in FireplaceQu
```

```python
# Replacing missing values of garage columns with 0s 
train_data['GarageYrBlt'] = train_data['GarageYrBlt'].fillna(0) 
train_data['GarageCond'] = train_data['GarageCond'].fillna(0)
train_data['GarageType'] = train_data['GarageType'].fillna(0)
train_data['GarageFinish'] = train_data['GarageFinish'].fillna(0)
train_data['GarageQual'] = train_data['GarageQual'].fillna(0)
```

```python
# Replacing missing values of basement columns with 0s 
train_data['BsmtFinType2'] = train_data['BsmtFinType2'].fillna(0) 
train_data['BsmtExposure'] = train_data['BsmtExposure'].fillna(0)
train_data['BsmtQual'] = train_data['BsmtQual'].fillna(0)
train_data['BsmtCond'] = train_data['BsmtCond'].fillna(0)
train_data['BsmtFinType1'] = train_data['BsmtFinType1'].fillna(0)
```

```python
# Replacing last missing values with 0s and mode
train_data['MasVnrArea'] = train_data['MasVnrArea'].fillna(0)
train_data['MasVnrType'] = train_data['MasVnrType'].fillna(0)
train_data['Electrical'].fillna(train_data['Electrical'].mode()[0], inplace=True)
```

```python
# Visualizing distribution of LotFrontage missing values to assess which imputation strategy to use 
import seaborn as sns

sns.boxplot(train_data.LotFrontage)
```

```python
# Replacing missing values of LotFrontage with median
train_data['LotFrontage'].fillna((train_data['LotFrontage'].median()), inplace=True)
```

```python
# Dropping unnecessary columns
train_data = train_data.drop(['PoolQC', 'PoolArea', 'MiscFeature', 'Alley', 'Fence'], axis=1) # dropping also Pool Area since the majority of values are equal to 0
```

```python
print(train_data.isnull().sum())
```

# Feature Engineering

```python
# Combining the two datasets
combined = [train_data, test_data]
```

### Neighborhood

```python
# Converting the categorical feature of Neighborhood to numeric
for dataset in combined:
    dataset["Neighborhood"] = dataset["Neighborhood"].map({"NAmes": 0, "CollgCr": 1, "OldTown": 2, "Edwards": 3, "Somerst": 4, "Gilbert": 5, "NridgHt": 6, "Sawyer": 7, "NWAmes": 8, "SawyerW": 9, "BrkSide": 10, "Crawfor": 11, "Mitchel": 12, "NoRidge": 13, "Timber": 14, "IDOTRR": 15, "ClearCr": 16, "StoneBr": 17, "SWISU": 18, "MeadowV": 19, "Blmngtn": 20, "BrDale": 21, "Veenker": 22, "NPkVill": 23, "Blueste": 24}).astype(int)
```

### Electrical

```python
# Converting the categorical feature of Electrical to numeric
for dataset in combined:
    dataset["Electrical"] = dataset["Electrical"].map({"SBrkr": 0, "FuseA": 1, "FuseF": 2, "FuseP": 3, "Mix": 4}).astype(int)
```

### Years Since Built

```python
# Calculating house age
current_year = 2023

# Adding a house age column
for dataset in combined:
    dataset['YearsSinceBuilt'] = current_year - dataset['YearBuilt']
```

### Years Since Sold

```python
# Calculating years since sold
for dataset in combined:
    dataset['YearsSinceSold'] = current_year - dataset['YrSold']
```

# Training our Machine Learning Model

```python
# Determining the input variable
X_train = train_data.drop("SalePrice", axis=1)
X_features = ['Neighborhood', 'YearsSinceBuilt', 'YearsSinceSold', 'OverallQual', 'Electrical']
```

```python
# Determining the output variable
Y_train = train_data['SalePrice']
```

```python
# Importing the Random Forest Classifier
from sklearn.ensemble import RandomForestClassifier
```

```python
# Creating an instance of the Random Forest Classifier
model = RandomForestClassifier()
```

```python
# Fitting the model using the training data
model.fit(X_train[X_features], Y_train)
```

# Testing our Model 

```python
# Reassigning the test dataset to the variable X_test
X_test = test_data
```

```python
# Predicting the features of the test set
predictions = model.predict(X_test[X_features])
```

```python
output = pd.DataFrame({'Id': test_data.Id, 'SalePrice': predictions})
output.to_csv('submission.csv', index=False)
print("Your submission was successfully saved!")
```
